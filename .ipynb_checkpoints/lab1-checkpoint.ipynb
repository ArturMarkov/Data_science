```python
def data_delimer(line):
    if '/' in line:        
        return ''    
    line = line.replace(' ', ',',2)    
    return (line + '\n')
```

```python
def clear_dir(directory, index):
    for file in glob.glob(os.path.join(directory, f"province-{index}*.csv")):
        os.remove(file) #delete all previous data
    
def download_data(directory, index, minYear=1991, maxYear=2020):    
    with requests.Session() as sess:
        url = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_provinceData.php?country=UKR&provinceID={INDEX[index]}&year1={minYear}&year2={maxYear}&type=Mean"
        response = sess.get(url)
    time =datetime.now().strftime("%d-%m-%Y_%H-%M")            
    path = os.path.join(directory, f"province-{index}.{time}.csv")        
            
    with open(path, 'w') as file:
        file.write("Year,Week,SMN,SMT,VCI,TCI,VHI\n")        
        for line in response.iter_lines(chunk_size=512, decode_unicode=True):                                            
            file.write(data_delimer(line)) #write line by line
```

```python
for i in range(1,26):
    clear_dir("D:\python_data_science\lab1\data", i)
    download_data("D:\python_data_science\lab1\data", i)
    print(f"Province {i} has been downloaded")
```

```python
def load_all_data_to_pd(path):
    df = pd.DataFrame(columns=['Year','Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI'])
    for file in glob.glob(os.path.join(path,"province-*.csv")):        
        temp = pd.read_csv(file, header=0)
        #specify province
        temp['Province'] = file.split('-')[1].split(".")[0] 
        df = df.append(temp, ignore_index=True)
    
       #change types for memory saving
        df['Province'] = df['Province'].astype("int32")
    #chagne colums consequence
    df = df[['Year','Week','Province','SMN', 'SMT', 'VCI', 'TCI', 'VHI']]
    
    
    
    return df
```

```python
def vhi_per_year(year):
    """
    Ряд VHI для області за рік, пошук екстремумів (min та max)
    """
    #print(year)
    
    return df[df.Year==year].set_index('Province')['VHI'] 
def vhi_per_province(province):
    return df[df.Province==province].set_index('Year')['VHI']
    
        
def min_max_vhi_per_province(province,flagmin, flagmax):
    t = vhi_per_province(province)
    if flagmin:
        return t.min()
    if flagmax: 
        return t.max()
    else:
        return "err"



def vhi_per_year_and_province(year,province):
   
    return df[(df.Year==year)&(df.Province==province)].set_index('Year','Province')['VHI']
   

    
    
def vhi_per_province_and_persent(province,persent,vhi_min=0,vhi_max=15):
    """Ряд VHI за всі роки для області, виявити  роки з екстремальними посухами,
    які торкнулися більше вказаного відсотка області"""
    years = df[(df['Province'] == province)]['Year'].unique()
    output=[]
    for year in years:
        df1=vhi_per_year_and_province(year,province)
        part =df1[(df1 >= vhi_min) & (df1 <= vhi_max)].count() * 100 /df1.count()
        if part>=persent:

            output.append(year)

    return output 
           
    
    
```
